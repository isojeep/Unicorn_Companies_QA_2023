## Exploratory Data Analyisis: Relationship, Insights and Visualisation

#### Univariate, Bivariate & Multivariate Analysis

```python
#Cities by Unicorn Companies - Top 5
cities_top = unicorn_data['City'].value_counts().sort_values(ascending= False)
cities_top = unicorn_data.groupby(['City']).size().reset_index(name='counts').sort_values(by=['counts'], ascending=False).head(5)
cities_top
```

```python
sns.set(style="whitegrid")
ax = sns.barplot(x="counts", y="City", data=cities_top)
plt.title('Top 5 Cities with the Most Unicorn Companies')
plt.xlabel('Number of Unicorn Companies')
plt.ylabel('City')

for i in ax.containers:
    ax.bar_label(i, label_type='edge', fontsize=10)

plt.show()
```

```python
#Countries by Unicorn Companies - Top 5
countries_top = unicorn_data['Country'].value_counts().sort_values(ascending= False)
countries_top = unicorn_data.groupby(['Country']).size().reset_index(name='counts').sort_values(by=['counts'], ascending=False).head(5)

#Visualisation of Contries by Unicorn Companies
sns.set(style="whitegrid")
ax = sns.barplot(x="counts", y="Country", data=countries_top)
plt.title('Top 5 Country with the Most Unicorn Companies')
plt.xlabel('Number of Unicorn Companies')
plt.ylabel('Country')

for i in ax.containers:
    ax.bar_label(i, label_type='edge', fontsize=10)

plt.show()
```

```python
#Companies by Valuation - Top 5
companies_top = unicorn_data.sort_values(by='Valuation', ascending=False).head(5)
companies_top
```

```python
#Visualisation - Companies by Valuation
sns.set(style="whitegrid")
plt.figure(figsize=(10,5))
ax = sns.barplot(x="Company", y="Valuation", data=companies_top)
ax.set_title('Top 5 Companies by Valuation')
ax.set_xlabel('Company')
ax.set_ylabel('Valuation (in billions of dollars)')
for i in ax.containers:
    ax.bar_label(i, label_type='edge', fontsize=10)

plt.show()
```

```python
#Companies by funding - Top 5
unicorn_data['Funding'] = unicorn_data['Funding'].astype(str).replace('Unknown', '0').str.replace('$','').str.replace('B','').str.replace('M','').fillna(0).astype(int)
companies_top = unicorn_data.sort_values(by='Valuation', ascending=False).head(5)
companies_top = companies_top.sort_values(by='Funding', ascending=True)

plt.barh(companies_top['Company'], companies_top['Funding'], color=['green', 'red', 'purple', 'orange', 'blue'])
plt.title('Top 5 Companies by Funding')
plt.xlabel('Funding (in billions of dollars)')

plt.ylabel('Company')

plt.show()
```

```python
unicorn_data['Funding'] = unicorn_data['Funding'].astype(str).replace('Unknown', '0').str.replace('$','').str.replace('B','').str.replace('M','').fillna(0).astype(int)

companies_top = unicorn_data.sort_values(by='Valuation', ascending=False).head(5)

plt.pie(companies_top['Funding'], labels=companies_top['Company'], autopct='%1.1f%%')
plt.title('Top 5 Companies by Funding')
plt.show()
```

```python
#Years in Unicorn - Top 5
# Convert the 'Date Joined' column to datetime
companies_top['Date Joined'] = pd.to_datetime(companies_top['Date Joined'])

# Calculate the number of years it took for each company to join the unicorn club
companies_top['Years to Unicorn'] = (companies_top['Date Joined'] - companies_top['Year Founded']).dt.days / 365

companies_top
```

```python
# Visualisation of Years in Unicorn - Top 5 Line Chart
# Create a line chart
sns.set(style="whitegrid")
plt.figure(figsize=(10,5))
ax = sns.lineplot(data=companies_top, x='Company', y='Years to Unicorn')
ax.set_title('Years to Unicorn for Top 5 Companies')
ax.set_xlabel('Company')
ax.set_ylabel('Years to Unicorn')

plt.show()
```

```python
# Visualisation - Years in Unicorn - Top 5 Bar Chart
# Create a bar chart
sns.set(style="whitegrid")
plt.figure(figsize=(10,5))
ax = sns.barplot(data=companies_top, x='Company', y='Years to Unicorn')
ax.set_title('Years to Unicorn for Top 5 Companies')
ax.set_xlabel('Company')
ax.set_ylabel('Years to Unicorn')

for i in ax.containers:
    ax.bar_label(i, label_type='edge', fontsize=10, fmt='%.1f')

plt.show()
```

```python
# Split the values in the 'Select Investors' column by comma
investors = unicorn_data['Select Investors'].str.split(', ')

# Explode the resulting Series to create a row for each investor
investors = investors.explode()

# Count the occurrences of each investor
s_investors = investors.value_counts().sort_values(ascending=False)

# Display the resulting Series
s_investors
```

```python
# Select the first five investors with the highest count
top_investors = s_investors.head()

# Define a list of colors for the bars
colors = ['red', 'orange', 'yellow', 'green', 'blue']

# Create a bar chart of the top investors with different colors for the bars
ax = top_investors.plot(kind='bar', color=colors)

# Add labels to the bars
for i in ax.containers:
    ax.bar_label(i)

# Add labels and title
plt.xlabel('Select Investors')
plt.ylabel('Number of Investments')
plt.title('Top 5 Select Investors Internationally')
plt.xticks(rotation=80)
# Show the plot
plt.show()
```

```python
# Select the first five investors with the highest count
top_investors = s_investors.head()

# Create a pie chart of the top investors
top_investors.plot(kind='pie', autopct='%1.1f%%')

# Add title
plt.title('Top 5 Unicorn Company Investors Internationally')
plt.ylabel('')

# Show the plot
plt.show()
```
